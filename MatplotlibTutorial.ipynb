```python
import csv
with open('data.csv') as csv_file:
    # Makes a dictionary where you can access by key rather than index
    csv_reader = csv.DictReader(csv_file)
    language_counter = Counter()
    
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
        
print(language_counter)
```

```python
# Print top 15
# Returns list with tuples
print(language_counter.most_common(15))
```

```python
# Split languages and counts to separate lists

# longer version
languages = []
popularity = []
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print()
print(popularity)
```

```python
# Horizontal might be easier to read
# Expects y-axis first
# Make sure to switch labels

# Can reverse lists to put the most popular at the top
languages.reverse()
popularity.reverse()

# Horizontal Barchart
plt.barh(languages, popularity)
plt.title('Most Popular Languages')
plt.ylabel('Programming Language')
plt.xlabel('Number of Users');
```

```python
data = pd.read_csv('data2.csv')
# Grabs columns with data from csv file
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, dev_salaries, color='#444444', linestyle='--', label='All Devs')

plt.plot(ages, py_salaries, label='Python')

overall_median = 57287

# Create fill
# Can add extra parameters (y2) which is set to 0 by default
# Alpha is the transparency
# Where is fill condition
# Interpolate makes sure x intersections do not get clipped and regions are filled correctly
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries > overall_median), interpolate=True, alpha=0.25)

# Another fill
# Can change color with 'color='
plt.fill_between(ages, py_salaries, overall_median, where=(py_salaries <= overall_median), interpolate=True, alpha=0.25, label='Above Average')

# Fill between lines (devsalaries as second boundary)
# 'label=' labels the filled area
plt.fill_between(ages, py_salaries, dev_salaries, where=(py_salaries > overall_median), interpolate=True, alpha=0.25)

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
```

```python
# Generate date into csv file
import csv
import random
import time

# Start values
x_value = 0
total_1 = 1000
total_2 = 1000

# Headers
fieldnames = ["x_value", "total_1", "total_2"]

with open('data5.csv', 'w') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()

# Continuously run
while True:

    with open('data5.csv', 'a') as csv_file:
        csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

        # Write out data for row
        info = {
            "x_value": x_value,
            "total_1": total_1,
            "total_2": total_2
        }

        csv_writer.writerow(info)
        print(x_value, total_1, total_2)

        x_value += 1
        total_1 = total_1 + random.randint(-6, 8)
        total_2 = total_2 + random.randint(-5, 6)

    time.sleep(1)
```

```python
# Called stateful method
# Figures: container holding different plots
# Axes: actual plots
# A figure can have multiple plots
# Previous examples are one plot within one figure (one axis)

data = pd.read_csv('data6.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# Takes in number of rows and columns (default is 1x1)
# Returns list objects of plots
# Can unpack ax as (ax1, ax2) in variable assignment to choose which ones to plot
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, sharex=True, sharey=True)

# To plot separately (two different figures rather then one figure with multiple axes)
# fig1, ax1 = plt.subplots()
# fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
# Add set_ before the labels
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

# To save the figures in current directory
# fig1.savefig('fig1.png')
# fig2.savefig('fig2.png')
```
